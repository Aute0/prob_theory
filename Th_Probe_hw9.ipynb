# **Задача №1**

## C intercept

Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):

zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],

ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].

* Ковариация: cov = 9157.84 

* Коэффициент корреляции Пирсона: r = 0.89

* Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). 
* Произвести расчет как с использованием intercept, так и без.

Модель выбираем линейную:

$$\hat{ks} = β_0 + β_1 * zp$$

Найдем коэффициенты β

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
b1 = (np.mean(zp * ks) - np.mean(zp) * np.mean(ks)) / \
    (np.mean(zp**2) - np.mean(zp)**2)
b0 = np.mean(ks) - b1 * np.mean(zp)
print(f"b0 = {b0} - intercept\nb1 = {b1}")
```

```python
# Python Check
model = LinearRegression()

zp = zp.reshape(-1, 1)

regres = model.fit(zp, ks)

b0 = regres.intercept_
b1 = regres.coef_[0]

print(f"b0 = {b0} - intercept\nb1 = {b1}")
```

Раасчеты яерез формулу линейной регрессии и код сошлись. Теперь произведем рассчет прогнозируемого значения скоринга.

```python
ks_pred_intercept = 444.1773573243595 + 2.6205388824027662 * zp
ks_pred_intercept
```

```python
ks_pred_intercept = model.predict(zp)
ks_pred_intercept
```

```python
plt.scatter(zp, ks, color = 'green', marker='*')
plt.title('Зависимость поведенческого кредитного скоринга от заработной платы')
plt.plot(zp, ks_pred_intercept, label=r'ks = 444.18 + 2.62 * zp',
         color='black', ls='dotted')
plt.legend()
plt.xlabel('Заработная плата заемщиков ')
plt.ylabel('Поведенческий кредитный скоринг')
plt.show()
```

## Без intercept

Без intercept будем использовать простейшую модель

$$\hat{ks} = β_1 * zp$$

Величину коэффициена b расчитываем с помощью матричного метода

```python
# независимая zp
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
# массив --> в-р
ks = ks.reshape((-1, 1))
zp = zp.reshape((-1, 1))

B = np.dot(np.linalg.inv(np.dot(zp.T, zp)), np.dot(zp.T, ks))[0, 0]
print(f"b1 = {B}")
```

Проверим тот же коеффициент с помощью встроеннных функций Python:

```python
# Python Check
model = LinearRegression(fit_intercept=False)
zp = zp.reshape(-1, 1)
regres = model.fit(zp, ks)

b0 = regres.intercept_
b1 = regres.coef_[0, 0]

print(f"b0 = {b0} - intercept\nb1 = {b1}")
```

## Произведем рассчет прогнозируемых значений:

```python
ks_pred = 5.889820420132688 * zp
ks_pred
```

```python
ks_pred = model.predict(zp)
ks_pred
```

```python
plt.scatter(zp, ks, color = 'green', marker='*')
plt.title('Зависимость поведенческого кредитного скоринга от заработной платы')
plt.plot(zp, ks_pred_intercept, label=r'ks = 444.18 + 2.62 * zp', color = 'black')
plt.plot(zp, ks_pred, label=r'ks = 5.89 * zp')
plt.legend()
plt.xlabel('Величина заработной платы заемщиков банка(zp)')
plt.ylabel('Поведенческий кредитный скоринг (ks)')
plt.show()
```

## Произведем оценку полученной модели:

```python
model = LinearRegression(fit_intercept=False)

zp = zp.reshape(-1, 1)

regres = model.fit(zp, ks)
regres.score(zp, ks)
```

```python
model = LinearRegression()

zp = zp.reshape(-1, 1)

regres = model.fit(zp, ks)
regres.score(zp, ks)
```

**Ответ:** Было получено, что предиктиивная модель с intercept выдает значения максимально близкие к фактическим, в отличие от модели без intercept, которая, в свою очередь показала результаты хуже, чем усредненные данные.


# **Задача №2**

Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

**Решение:** воспользуемся формулой:

$$\hat{ks} = β_1 * zp$$

```python
# признак 
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
# функция потерь для B1
def mse_(B1, y=y, x=x, n=len(y)):
  return np.sum((B1*x - y)**2) / n
```

```python
alpha = 0.000001  # скорость обучения как в лекции
B1 = 0.1
n = len(y)
```

```python
# каждую 500-ую итерацию
for i in range(2000):
  B1 -= alpha * (2/n) * np.sum((B1*x - y) * x)
  if i % 500 == 0:
    print(f"Итерация {i}: B1 = {B1}, mse = {mse_(B1)}")
```

**Ответ:** без intercept с помощью метода градиентного спуска коэффициэнт b = 5.889820420132673, что соответствует предыдущим результатам.

# **Задача №3**

Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

**Решение:** 

воспользуемся формулой:
$$\hat{ks} = β_0 + β_1 * zp$$

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
def _mse(b0, b1, x=zp, y=ks):
    return np.sum(((b0 + b1 * x) - y)**2) / len(x)
```

```python
def _mse_b0(b0, b1, x = zp, y = ks): 
    return 2 * np.sum((b0 + b1 * x) -y) / len(x)
```

```python
def _mse_b1(b0, b1, x = zp, y = ks):
    return 2 * np.sum(((b0 + b1 * x) - y) * x) / len(x)
```

```python
alpha = 5e-05
b1 = 0.1
b0 = 0.1
mse_min = _mse(b0, b1)
i_min = 1
b1_min = b1
b0_min = b0

for i in range(1000000):
    b0 -= alpha * _mse_b0(b0, b1)
    b1 -= alpha * _mse_b1(b0, b1)
    if i % 50000 == 0:
        print(f'Итерация {i}: b0 = {b0}, b1 = {b1}, mse = {_mse(b0, b1)}')
    if _mse(b0, b1) > mse_min:
        print(
            f'Итерация {i_min}: b0 = {b0_min}, b1 = {b1_min}, mse = {mse_min},\nДостигнут минимум.')
        break
    else:
        mse_min = _mse(b0, b1)
        i_min = i
        b1_min = b1
        b0_min = b0
print(f'b0 = {b0_min} \nb1 = {b1_min}')
```
